# Data Preparation

## Join Patient and Diagnosis

```python
merged_df = pd.merge(patient, diagnosis, on='patientunitstayid', how='inner')
```

```python
merged_df
```

```python
ft = merged_df.copy()
```

```python
ft.head()
```

## เลือกเฉพาะผู้ป่วยที่เข้ามารักษาที่โรงพยาบาลเกิน 1 วัน

```python
p_t = ft[['patientunitstayid','hospitaladmitoffset','unitdischargeoffset','hospitaladmittime24','unitdischargetime24','hospitaldischargeoffset','age','diagnosisid','activeupondischarge','diagnosisoffset','diagnosisstring']]
```

```python
p_t['offset_difference'] = p_t['unitdischargeoffset'] - p_t['hospitaladmitoffset']
```

```python
p_t_m = p_t[p_t['offset_difference'] > 1440 ]
p_t_m
```

```python
merged_1d = p_t_m[['patientunitstayid','age','offset_difference','diagnosisid','activeupondischarge','diagnosisoffset','diagnosisstring']]
merged_1d
```

## เลือกเฉพาะผู้ป่วยที่มีอายุ 18 ปีขึ้นไป

```python
ft_non89 = merged_1d[merged_1d['age'] != '> 89']
```

```python
ft_89 = merged_1d[merged_1d['age'] == '> 89']
```

```python
ft_non89['age'] = pd.to_numeric(ft_non89['age'], errors='coerce')
```

```python
ft_18 = ft_non89[ft_non89['age'] >= 18]
```

```python
ft_18
```

```python
merged_1d_18 = pd.concat([ft_18, ft_89])
merged_1d_18
```
