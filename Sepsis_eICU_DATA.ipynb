### แปลงเป็นตารางที่พร้อมจะนำไป train model
- ทำ timestamp ให้มี timeline ตรงกัน

```python
data = sepsis_group_lab_12.copy()
```

```python
# Check the columns in the data
print(data.columns)

# Assuming the file has columns: ['patientunitstayid', 'labname', 'labresultoffset', 'labresult']
# Convert 'labresultoffset' from minutes to hours
data['labresultoffset'] = data['labresultoffset'] / 60

# Round the time to the nearest hour
data['labresultoffset'] = data['labresultoffset'].round().astype(int)

# Set the start time for each patient to the first recorded data point
data['labresultoffset'] = data.groupby('patientunitstayid')['labresultoffset'].transform(lambda x: x - x.min())

# Aggregate duplicates by averaging them
data = data.groupby(['patientunitstayid', 'labname', 'labresultoffset']).agg('mean').reset_index()

# Create an hourly time range for each patient and lab name
def reindex_group(group):
    idx = pd.Index(range(group['labresultoffset'].min(), group['labresultoffset'].max() + 1), name='labresultoffset')
    reindexed_group = group.set_index('labresultoffset').reindex(idx).reset_index()
    reindexed_group['patientunitstayid'] = group['patientunitstayid'].iloc[0]
    reindexed_group['labname'] = group['labname'].iloc[0]
    return reindexed_group

aligned_data = data.groupby(['patientunitstayid', 'labname']).apply(reindex_group).reset_index(drop=True)

# Forward-fill missing values
aligned_data['labresult'] = aligned_data.groupby(['patientunitstayid', 'labname'])['labresult'].ffill()

# Apply mean imputation for entirely missing features
grouped = aligned_data.groupby(['patientunitstayid', 'labname'])
aligned_data['labresult'] = aligned_data['labresult'].astype('float64')
aligned_data['labresult'] = grouped['labresult'].transform(lambda x: x.fillna(x.mean()))
aligned_data
```

```python
pivoted_exam = aligned_data.pivot(index=['patientunitstayid', 'labname'], columns='labresultoffset', values='labresult')
```

```python
# Function to forward-fill and fill mean conditionally
def conditional_fill(df):
    for col in df.columns[1:]:
        prev_col = df[col].shift(1)
        threshold_60 = prev_col * 0.6

        # Condition for forward-fill
        condition_ffill = df[col].isna() & (prev_col < threshold_60)
        # Condition for fill mean
        condition_mean = df[col].isna() & (prev_col >= threshold_60)

        # Forward-fill based on the condition
        df.loc[condition_ffill, col] = df.loc[condition_ffill].ffill(axis=1)

        # Fill mean based on the condition
        row_means = df.mean(axis=1)
        df.loc[condition_mean, col] = df.loc[condition_mean].apply(lambda row: row_means[row.name], axis=1)

    return df

# Apply the conditional forward-fill and mean fill
pivoted_exam_filled = conditional_fill(pivoted_exam)
```

```python
pivoted_exam_filled = pivoted_exam_filled.fillna(pivoted_exam_filled.mean(axis=0))
```
