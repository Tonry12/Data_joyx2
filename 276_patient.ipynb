```python
# Calculate percentage of NaN values for each column
nan_percentages = data.isnull().mean(axis=0) * 100

# Threshold for dropping columns
threshold = 40

# Get columns to drop (where NaN percentage is above threshold and column is not 'MAP' or 'BP_SYS')
columns_to_drop = [col for col in nan_percentages[nan_percentages >= threshold].index if col not in ['MAP', 'BP_SYS']]

# Drop the selected columns
data.drop(columns_to_drop, axis=1, inplace=True)
```
