```python
import pandas as pd
from datetime import timedelta
df['ICUDateTime'] = pd.to_datetime(df['ICUDateTime'])

# Determine the time of sepsis based on the condition
df['Sepsis'] = df.apply(lambda row: row['ICUDateTime'] if (row['MAP'] < 65 or row['BP_SYS'] < 90) else None, axis=1)

# Drop rows where Sepsis is not detected
df_sepsis = df.dropna(subset=['Sepsis'])

# Find the earliest sepsis time for each AdmissionID
earliest_sepsis_times = df_sepsis.groupby('AdmissionID')['Sepsis'].min()

# Create a column to hold the time 1 hour before sepsis
df['SepsisMinus1Hour'] = df['AdmissionID'].map(lambda x: earliest_sepsis_times[x] - timedelta(hours=1) if x in earliest_sepsis_times else None)

# Filter the rows where ICUDateTime is within 1 hour before sepsis
filtered_df = df[df.apply(lambda row: row['ICUDateTime'] >= row['SepsisMinus1Hour'] if row['SepsisMinus1Hour'] is not None else False, axis=1)]

# Drop the auxiliary columns
filtered_df = filtered_df.drop(columns=['Sepsis', 'SepsisMinus1Hour'])
filtered_df
```

```python
# Convert ICUDateTime to datetime
df['ICUDateTime'] = pd.to_datetime(df['ICUDateTime'])

# Determine the time of sepsis based on the condition
df['Sepsis'] = df.apply(lambda row: row['ICUDateTime'] if (row['MAP'] < 65 or row['BP_SYS'] < 90) else None, axis=1)

# Drop rows where Sepsis is not detected
df_sepsis = df.dropna(subset=['Sepsis'])

# Find the earliest sepsis time for each AdmissionID
earliest_sepsis_times = df_sepsis.groupby('AdmissionID')['Sepsis'].min()

# Create columns to hold the time 1 hour before sepsis and 5 hours before sepsis
df['SepsisMinus1Hour'] = df['AdmissionID'].map(lambda x: earliest_sepsis_times[x] - timedelta(hours=1) if x in earliest_sepsis_times else None)
df['SepsisMinus5Hours'] = df['AdmissionID'].map(lambda x: earliest_sepsis_times[x] - timedelta(hours=2) if x in earliest_sepsis_times else None)

# Filter the rows where ICUDateTime is within 1 to 5 hours before sepsis
filtered_df = df[df.apply(lambda row: row['SepsisMinus5Hours'] <= row['ICUDateTime'] < row['SepsisMinus1Hour'] if row['SepsisMinus1Hour'] is not None and row['SepsisMinus5Hours'] is not None else False, axis=1)]

# Drop the auxiliary columns
filtered_df = filtered_df.drop(columns=['Sepsis', 'SepsisMinus1Hour', 'SepsisMinus5Hours'])
filtered_df
```
