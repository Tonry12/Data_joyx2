```python
import pandas as pd

# Sample data
data = {
    'SETTING_ModeVentilator': [
        'A/C PC', 'CPAP/PSV', 'A/C VC', 'O2 Therapy', 'A/C PRVC', 'NIV',
        'SIMV-PC', 'A/CMV-PC', 'SPONT-PS', 'VS', 'O2 Trp.', 'BiLevel',
        'O2 Trp', 'PC/AC', 'O2 therapy', 'HFNC'
    ]
}

# Create DataFrame
completed = pd.DataFrame(data)

# Mapping dictionary
mapping = {
    'A/C PC': 1,
    'CPAP/PSV': 2,
    'A/C VC': 3,
    'O2 Therapy': 4,
    'A/C PRVC': 5,
    'NIV': 6,
    'SIMV-PC': 7,
    'A/CMV-PC': 8,
    'SPONT-PS': 9,
    'VS': 10,
    'O2 Trp.': 11,
    'BiLevel': 12,
    'O2 Trp': 13,
    'PC/AC': 14,
    'O2 therapy': 15,
    'HFNC': 16
}

# Replace values
completed['SETTING_ModeVentilator'] = completed['SETTING_ModeVentilator'].map(mapping)
```
