### admission

```python
admission = pd.read_csv(path+"icu_admission.csv")
admission = admission.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
admission.drop(['Unnamed: 0'], axis=1, inplace=True)
admission.VN = admission.VN.astype('Int64')
admission.AdmissionID = admission.AdmissionID.astype('Int64')
admission.info(verbose=True)
```

```python
print(admission['HN'].nunique(), admission['AN'].nunique())
```

```python
admission.sort_values(by=['DateRecord','TimeRecord'], ascending=True, inplace=True)
admission = admission.drop_duplicates(subset=['HN'], keep='first')
```

```python
print(admission['HN'].nunique(), admission['AN'].nunique())
```

# convert `ICUDate` and `ICUTime` to datetime format

```python
admission['ICUDateTime'] = pd.to_datetime([str(i)+' '+str(j) for i,j in zip(admission['DateRecord'],admission['TimeRecord'])], format='%Y-%m-%d %H:%M')
bedccr['ICUDateTime'] = pd.to_datetime([str(i)+' '+str(j) for i,j in zip(bedccr['ICUDate'],bedccr['ICUTime'])], format='%d-%m-%Y %H')
bedccr_lab['ICUDateTime'] = pd.to_datetime([str(i)+' '+str(j) for i,j in zip(bedccr_lab['ICUDate'],bedccr_lab['ICUTime'])], format='%d-%m-%Y %H')
bedmonitortrack['ICUDateTime'] = pd.to_datetime([str(i)+' '+str(j) for i,j in zip(bedmonitortrack['ICUDate'],bedmonitortrack['ICUTime'])], format='%d-%m-%Y %H')
```

```python
redcap['h_adm_date'] = pd.to_datetime(redcap['h_adm_date'], format='%Y-%m-%d')
redcap['icu_adm_date'] = pd.to_datetime(redcap['icu_adm_date'], format='%Y-%m-%d')
admission['DateRecord'] = pd.to_datetime(admission['DateRecord'], format='%Y-%m-%d')
admission['TimeRecord'] = pd.to_datetime(admission['TimeRecord'], format='%H:%M')
bedccr['ICUDate'] = pd.to_datetime(bedccr['ICUDate'], format='%d-%m-%Y')
bedccr['ICUTime'] = pd.to_datetime(bedccr['ICUTime'], format='%H')
bedccr_lab['ICUDate'] = pd.to_datetime(bedccr_lab['ICUDate'], format='%d-%m-%Y')
bedccr_lab['ICUTime'] = pd.to_datetime(bedccr_lab['ICUTime'], format='%H')
bedmonitortrack['ICUDate'] = pd.to_datetime(bedmonitortrack['ICUDate'], format='%d-%m-%Y')
bedmonitortrack['ICUTime'] = pd.to_datetime(bedmonitortrack['ICUTime'], format='%H')
intake_output['ICUDate'] = pd.to_datetime(intake_output['ICUDate'], format='%d-%m-%Y')
```
