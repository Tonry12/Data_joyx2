```python
filtered_df = merged_df
```

## เลือกเฉพาะผู้ป่วยที่มาอายุ 18+

```python
ft = filtered_df[filtered_df['diagnosisstring'].str.contains('sepsis', case=False, na=False)]
```

```python
ft
```

```python
import numpy as np
np.unique(ft['age'])
```

```python
ft_non89 = ft[ft['age'] != '> 89']
```

```python
ft_89 = ft[ft['age'] == '> 89']
```

```python
import numpy as np
len(np.unique(ft['patienthealthsystemstayid']))
```

```python
(np.unique(ft['patienthealthsystemstayid'])).size
```

```python
ft_non89['age'] = pd.to_numeric(ft_non89['age'], errors='coerce')
```

```python
ft_18 = ft_non89[ft_non89['age'] >= 18]
```

```python
ft_18
```

```python
ft_18_c = pd.concat([ft_18, ft_89])
```

```python
ft_18_c['hospitaldischargeoffset'] - ft_18_c['hospitaladmitoffset']
```

## ลบ missing ที่มากกว่า 30%

```python
ft_d = ft_18_c[(ft_18_c['hospitaldischargeoffset'] - ft_18_c['hospitaladmitoffset']) > 1440]
```

```python
missing_percentage = ft_d.isnull().mean(axis=1) * 100
```

```python
df_d = ft_d[missing_percentage <= 30]
```

```python
df_d
```

```python
df_d.dtypes
```

## ทำการหาเวลารวมของผู้ป่วยที่เข้ามารักษาในรพ. จนออกจาก รพ. เฉพาะเข้า ICU ครั้งแรก

```python
p_t = df_d[['patientunitstayid','hospitaladmitoffset','unitdischargeoffset','hospitaladmittime24','unitdischargetime24','hospitaldischargeoffset']]
```

```python
p_t['offset_difference'] = p_t['unitdischargeoffset'] - p_t['hospitaladmitoffset']
```

```python
p_t['offset_difference_transpost'] = p_t['hospitaladmitoffset'] + p_t['unitdischargeoffset']
```
