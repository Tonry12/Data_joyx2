```python
lotto.rename(columns={'เลข 4 ตัว': 'top4','เลข 3 ตัว':'top3','เลข 2 ตัว':'top2'},inplace=True)
```

```python
lt = lotto.drop(columns=['weekday_dum'])
```

```python
lt['date'] = pd.to_datetime(lt['date'])
```

```python
lt.dtypes
```

```python
lt['top4'] = lt['top4'].apply(lambda x: '{:04}'.format(x))
```

```python
lt
```

```python
lt_1 = lt
```

```python
lt_1['top4'] = lt_1['top4'].astype(str)

# Split each value in the 'digi' column into its individual digits
lt_1['r1'] = lt_1['top4'].apply(lambda x: int(x[0]))
lt_1['r2'] = lt_1['top4'].apply(lambda x: int(x[1]))
lt_1['r3'] = lt_1['top4'].apply(lambda x: int(x[2]))
lt_1['r4'] = lt_1['top4'].apply(lambda x: int(x[3]))
```

```python
lt_1= lt_1.drop(columns=['top3','top2'])
```

```python
lt_1
```

# วันจันทร์

## เลขท้าย 4 ตัว

```python
monday = lt_1[lt_1['weekday'] == "Monday"]
```

```python
monday.head()
```

```python
# เลขท้ายสี่ตัวหลักที่ 1 ที่ออกบ่อยสุด
grouped_counts = monday.groupby('r1').size().reset_index(name='Count')
max_count_row = grouped_counts[grouped_counts['Count'] == grouped_counts['Count'].max()]
max_count_row
```

```python
# เลขท้ายสี่ตัวหลักที่ 2 ที่ออกบ่อยสุด
grouped_counts = monday.groupby('r2').size().reset_index(name='Count')
max_count_row = grouped_counts[grouped_counts['Count'] == grouped_counts['Count'].max()]
max_count_row
```

```python
# เลขท้ายสี่ตัวหลักที่ 3 ที่ออกบ่อยสุด
grouped_counts = monday.groupby('r3').size().reset_index(name='Count')
max_count_row = grouped_counts[grouped_counts['Count'] == grouped_counts['Count'].max()]
max_count_row
```

```python
# เลขท้ายสี่ตัวหลักที่ 4 ที่ออกบ่อยสุด
grouped_counts = monday.groupby('r4').size().reset_index(name='Count')
max_count_row = grouped_counts[grouped_counts['Count'] == grouped_counts['Count'].max()]
max_count_row
```
